## 《人工智能与深度学习》期末报告要求说明
### 一、总体要求
本课程的期末报告要求应用本课程所学的深度学习技术，独立完成一个分析项目。

选题和数据不限，可以是实证应用，即基于真实数据应用深度学习模型；也可以是方法创新，提出改进深度学习模型。

严禁抄袭等学术不端行为。

### 二、报告内容说明
期末报告需以 Jupyter Notebook (.ipynb) 的形式提交。Notebook 应包含完整的分析说明、研究代码与结果。要求所有实验结果须可复现，注意代码规范以及清晰的注释。

实证应用类项目推荐包括但不限于以下内容：
1. 引言：
   - 阐述选题背景
   - 明确研究目标与问题定义
   - 简要介绍采用的方法
2. 数据说明：
    - 说明数据来源，导入包与数据，并划分训练集与测试集
    - 说明预处理方法，并描述数据的主要特征（描述统计或可视化）等；
3. 模型设计：
   - 阐述模型结构，使用 `nn.Module` 构建你的模型
   - 定义训练以及测试工具，选择优化器及损失函数
4. 模型结果分析
   - 模型训练与超参数调优，报告训练与测试集的性能指标
   - 可与传统的统计方法进行对比，说明你的深度学习模型是否优于传统方法，并简要分析原因
   - 对结果进行现实层面的描述解释
5. 总结：
   - 总结研究发现与主要结论
   - 讨论局限性，提出未来改进方向或潜在应用场景
